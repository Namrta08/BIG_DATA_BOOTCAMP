```python
import pymongo
```

```python

# Provide the mongodb localhost url to connect python to mongodb.
client = pymongo.MongoClient("mongodb://localhost:27017/neurolabDB")
```

```python
# Database Name
db = client["iNeuron"]
```

```python
# Collection Name
coll = db["Admission_details"]
```

```python
# import pandas
import pandas as pd
```

```python
# reaad csv
df = pd.read_csv("https://raw.githubusercontent.com/vigneshk/Admission-Dataset/master/Admission.csv")
```

```python
# converting to dict
my_data = df.to_dict(orient = "records")
```

```python
my_data
```

```python
# insert data into collection
coll.insert_many(my_data)
```

```python
# fetch records from collection
for idx, i in enumerate(coll.find()):
    print(f"{idx}: {i}")
```
